```python
import os
import json
import re
import google.generativeai as genai
from dotenv import load_dotenv

from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
```

```python
load_dotenv()

api_key = os.getenv("GEMINI_API_KEY")

genai.configure(api_key=api_key)
print(api_key)

model = genai.GenerativeModel("gemini-2.5-flash")

print("Gemini connected successfully")
```

```python
prompt = "Explain the difference between supervised and unsupervised learning in simple terms."

response = model.generate_content(prompt)

# Print the AI-generated text
print(response)
```

```python
kb_files = [
    "data/refund_policy.md",
    "data/account_upgrade.md",
    "data/api_rate_limits.md",
    "data/security_practices.md",
    "data/integration_setup.md"
]

documents = []

for file in kb_files:
    with open(file, "r", encoding="utf-8") as f:
        text = f.read()
        documents.append({"source": file, "content": text})

print("Documents loaded:", len(documents))
```

```python
embed_model = SentenceTransformer("all-MiniLM-L6-v2")

texts = [doc["content"] for doc in documents]

embeddings = embed_model.encode(texts)

dimension = embeddings.shape[1]

index = faiss.IndexFlatL2(dimension)

index.add(np.array(embeddings))

print("Vector database created")
```

```python
with open("data/tickets.json") as f:
    tickets = json.load(f)

with open("data/accounts.json") as f:
    accounts = json.load(f)

print("Tickets:", len(tickets))
print("Accounts:", len(accounts))
```

```python
def route_query(query):

    prompt = f"""
Classify the user query into one category:

KNOWLEDGE_BASE
TICKET_LOOKUP
ACCOUNT_LOOKUP
AMBIGUOUS
UNSUPPORTED

Rules:

KNOWLEDGE_BASE → documentation or policy questions
TICKET_LOOKUP → ticket id or ticket status questions
ACCOUNT_LOOKUP → customer plan or renewal questions
AMBIGUOUS → unclear request
UNSUPPORTED → not present in provided data

Return ONLY the category name.

Query:
{query}
"""

    response = model.generate_content(prompt)

    route = response.text.strip()

    return route
```

```python
def retrieve_docs(query, k=2):

    query_embedding = embed_model.encode([query])

    distances, indices = index.search(np.array(query_embedding), k)

    results = []

    for i in indices[0]:
        results.append(documents[i])

    return results
```

```python
def knowledge_answer(query):

    docs = retrieve_docs(query)

    context = ""

    sources = []

    for d in docs:
        context += d["content"] + "\n"
        sources.append(d["source"])

    prompt = f"""
Answer the question using ONLY the context below.

Context:
{context}

Question:
{query}
"""

    response = model.generate_content(prompt)

    return response.text, sources
```

```python
def ticket_lookup(query):

    match = re.search(r"T-\d+", query)

    if not match:
        return "Ticket ID not found.", ["tickets.json"]

    ticket_id = match.group()

    for t in tickets:

        if t["ticket_id"] == ticket_id:

            # Use get() to avoid KeyError
            agent = t.get("agent") or t.get("assigned_agent") or "Not assigned"

            answer = f"""
Ticket ID: {t['ticket_id']}
Status: {t.get('status', 'Unknown')}
Assigned Agent: {agent}
Priority: {t.get('priority', 'Unknown')}
Customer: {t.get('customer', 'Unknown')}
"""

            return answer, ["tickets.json"]

    return "Ticket not found.", ["tickets.json"]
```

```python
def account_lookup(query):

    for acc in accounts:

        name = acc["customer_name"].lower()

        if name in query.lower():

            answer = f"""
Customer: {acc['customer_name']}
Plan: {acc['plan']}
Renewal Date: {acc['renewal_date']}
Health Score: {acc['health_score']}
Open Tickets: {acc['open_ticket_count']}
"""

            return answer, ["accounts.json"]

    return "Account not found.", ["accounts.json"]
```

```python
def triage_priority():
    """
    Determines which tickets should be handled first based on priority and account health.
    Returns a sorted list of tickets and their scores.
    """
    results = []

    for t in tickets:
        score = 0

        # Basic priority scoring
        priority = t.get("priority", "").lower()
        if priority == "high":
            score += 3
        elif priority == "medium":
            score += 2
        elif priority == "low":
            score += 1

        # Safely get the account
        customer_name = t.get("customer")  # Use .get() to avoid KeyError
        acc = next((a for a in accounts if a["customer_name"] == customer_name), None)

        # Adjust score based on account health
        if acc:
            health_score = acc.get("health_score", 100)
            if health_score < 50:
                score += 2
            elif health_score < 80:
                score += 1

        # Save the ticket and its computed score
        results.append({
            "ticket_id": t.get("id"),
            "customer": customer_name,
            "summary": t.get("summary", "No summary provided"),
            "score": score
        })

    # Sort tickets by score descending (highest priority first)
    results.sort(key=lambda x: x["score"], reverse=True)

    # Return results and any extra info if needed
    return results, {"total_tickets": len(results)}
```

```python
def assistant(query):

    route = route_query(query)

    used_sources = []
    used_tools = []
    clarification = False
    confidence = 0.9

    if "handle first" in query.lower():

        answer, used_sources = triage_priority()

        route = "TRIAGE"

        used_tools = ["triage_logic"]

    elif route == "KNOWLEDGE_BASE":

        answer, used_sources = knowledge_answer(query)

        used_tools = ["rag_retriever"]

    elif route == "TICKET_LOOKUP":

        answer, used_sources = ticket_lookup(query)

        used_tools = ["ticket_lookup"]

    elif route == "ACCOUNT_LOOKUP":

        answer, used_sources = account_lookup(query)

        used_tools = ["account_lookup"]

    elif route == "AMBIGUOUS":

        answer = "Could you clarify your request?"
        clarification = True

    else:

        answer = "Sorry, that information is not available."

    decision = {
        "route": route,
        "confidence": confidence,
        "used_sources": used_sources,
        "used_tools": used_tools,
        "needs_clarification": clarification,
        "final_answer": answer
    }

    return answer, decision
```

```python
query = input("Ask a question: ")

answer, decision = assistant(query)

print("\nAnswer:\n")
print(answer)

print("\nDecision Object:\n")
print(json.dumps(decision, indent=2))
```
